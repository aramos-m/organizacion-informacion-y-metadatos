## <font color='FIREBRICK'>Ejercicio 1 (4 puntos)</font>

<font size='3px'><p style="text-align: justify;line-height: 150%">Queremos almacenar los datos de nuestros actores favoritos en una base de datos (BD), pero no nos gusta ninguna de las disponibles, por lo tanto vamos a construirla nosotros mismos.</p></font>
    
<font size='3px'><p style="text-align: justify;line-height: 150%">La BD será un diccionario de diccionarios. Para el primer diccionario la clave será el nombre del actor. El segundo diccionario (el que estaría dentro del primer diccionario) consta de 3 campos: año de nacimiento, sexo y una lista con el nombre de las películas en las que ha participado.</p></font>

<font size='3px'><p style="text-align: justify;line-height: 150%">El programa nos permitirá hacer una serie de cosas, elegidas por un menú:</p></font>

<div style="font-size: 16px; line-height: 150%">
<ol>
<li>Introducir los datos de un nuevo actor o actriz.
    <li>Listar todos los actores, especificando los 3 primeros campos.
    <li>Mostrar los datos de un determinado actor o actriz.
    <li>Buscar aquellos actores que se encuentren en una determinada franja de edad.
    <li>Buscar aquellos actores de un sexo determinado.
    <li>Buscar aquellos actores que hayan participado en una película.
</ol>
</div>

```python
# Realice el ejercicio dentro del cuadro a continuación de esta línea 
def menu_actores():
    BBDD_actores = {}

    while True:
        print("\n--- Menú actores ---")
        print("1 - Añadir un nuevo actor")
        print("2 - Listar todos los actores")
        print("3 - Ver datos de un actor")
        print("4 - Buscar por edad")
        print("5 - Buscar por sexo")
        print("6 - Buscar por película")
        print("7 - Salir")
        opcion = input("Introduzca una opción: ")
        
        if opcion == '1' :
            print("\nHa elegido la opción 1")
            nombre_actor = input("Introduzca el nombre del actor: ")
            anio = int(input("Introduzca su año de nacimiento: "))
            sexo = input("Introduzca su sexo (masculino/femenino): ")
            peliculas = input("Introduzca sus películas separadas por espacios: ").split()
    
            BBDD_actores[nombre_actor] = {'anio': anio, 'sexo': sexo, 'peliculas': peliculas}
            print("\n¡Actor añadido exitosamente!")
        
        elif opcion == '2':
            print("\nHa elegido la opción 2")
            for nombre_actor, datos_actor in BBDD_actores.items():
                print("Nombre:", nombre_actor, "; Año:", datos_actor['anio'], "; Sexo:", datos_actor['sexo'], "; Películas:", datos_actor['peliculas'])
        
        elif opcion == '3':
            print("\nHa elegido la opción 3")
            nombre_actor = input("Introduzca el nombre para buscar actores: ")
            if nombre_actor in BBDD_actores:
                print("Nombre:", nombre_actor, "; Año:", datos_actor['anio'], "; Sexo:", datos_actor['sexo'], "; Películas:", datos_actor['peliculas'])
            else:
                print("\nNo se encontraron actores.")
                
        elif opcion == '4':
            print("\nHa elegido la opción 4")
            edad_min = int(input("Introduzca la franja de edad para buscar actores:\n- Edad mínima: "))
            edad_max = int(input("- Edad máxima: "))
    
            actores_encontrados = False
            for nombre_actor, datos_actor in BBDD_actores.items():
                edad = 2024 - datos_actor['anio']
                if edad >= edad_min and edad <= edad_max:
                    actores_encontrados = True
                    print("Nombre:", nombre_actor, "; Año:", datos_actor['anio'], "; Sexo:", datos_actor['sexo'], "; Películas:", datos_actor['peliculas'])
            
            if actores_encontrados == False:
                print("\nNo se encontraron actores.")
                
        elif opcion == '5':
            print("\nHa elegido la opción 5")
            sexo = input("Introduzca el sexo para buscar actores: ").lower()
    
            actores_encontrados = False
            for nombre_actor, datos_actor in BBDD_actores.items():
                if datos_actor['sexo'] == sexo:
                    actores_encontrados = True
                    print("Nombre:", nombre_actor, "; Año:", datos_actor['anio'], "; Sexo:", datos_actor['sexo'], "; Películas:", datos_actor['peliculas'])
            
            if actores_encontrados == False:
                print("\nNo se encontraron actores.")

        elif opcion == '6':
            print("\nHa elegido la opción 6")
            nombre_pelicula = input("Introduzca la película para buscar actores: ")
        
            actores_encontrados = False
            for nombre_actor, datos_actor in BBDD_actores.items():
                if nombre_pelicula in datos_actor['peliculas']:
                    actores_encontrados = True
                    print("Nombre:", nombre_actor, "; Año:", datos_actor['anio'], "; Sexo:", datos_actor['sexo'], "; Películas:", datos_actor['peliculas'])
                    if nombre_pelicula in BBDD_peliculas:
                        datos_pelicula = BBDD_peliculas[nombre_pelicula]
                        print("Nombre:", nombre_pelicula, "; Año:", datos_pelicula['anio'], "; Género:", datos_pelicula['genero'], "; Duración:", datos_pelicula['duracion'])
            
            if actores_encontrados == False:
                print("\nNo se encontraron actores que participen en esa película.")
                        
        elif opcion == '7':
            print("\n¡Hasta luego!")
            break
            
        else:
            print("\nIntroduce una opción válida.")
```

## <font color='FIREBRICK'>Ejercicio 2 (2 puntos)</font>

<font size='3px'><p style="text-align: justify;line-height: 150%">Queremos almacenar los datos de nuestras películas favoritas en una base de datos (BD), pero no nos gusta ninguna de las disponibles, por lo tanto vamos a construirla nosotros mismos.</p></font>
    
<font size='3px'><p style="text-align: justify;line-height: 150%">La BD será una lista de diccionarios. Cada elemento del diccionario consta de 4 campos:
Nombre de la película, año de estreno, género y la duración.</p></font>

<font size='3px'><p style="text-align: justify;line-height: 150%">El programa nos permitirá hacer una serie de cosas, elegidas por un menú:</p></font>

<div style="font-size: 16px; line-height: 150%">
<ol>
    <li>Introducir los datos de una nueva película.</li>
    <li>Listar todas las películas, especificando todos los campos.</li>
    <li>Mostrar los datos de una determinada película.</li>
    <li>Buscar aquellas películas que se extrenasen en una determinada franja de años.</li>
    <li>Buscar aquellas películas de un género determinado.</li>
    <li>Buscar aquellas películas que tengan una duración próxima a la indicada.</li>
</ol>
</div>

```python
# Realice el ejercicio dentro del cuadro a continuación de esta línea  
def menu_peliculas():
    BBDD_peliculas = {}
    
    while True:
        print("\n--- Menú películas ---")
        print("1 - Añadir una nueva película")
        print("2 - Listar todas las películas")
        print("3 - Ver datos de una película")
        print("4 - Buscar por año")
        print("5 - Buscar por género")
        print("6 - Buscar por duración")
        print("7 - Salir")
        
        opcion = input("Introduzca una opción: ")
        
        if opcion == '1':
            print("\nHa elegido la opción 1")
            nombre_pelicula = input("Introduzca el nombre de la película: ")
            anio = int(input("Introduzca el año de la película: "))
            genero = input("Introduzca el género de la película: ")
            duracion = int(input("Introduzca la duración de la película en minutos: "))
    
            BBDD_peliculas[nombre_pelicula] = {'anio': anio, 'genero': genero, 'duracion': duracion}
            print("\n¡Película añadida exitosamente!")
        
        elif opcion == '2':
            print("\nHa elegido la opción 2")
            for nombre_pelicula, datos_pelicula in BBDD_peliculas.items():
                print("Nombre:", nombre_pelicula, "; Año:", datos_pelicula['anio'], "; Género:", datos_pelicula['genero'], "; Duración:", datos_pelicula['duracion'])
        
        elif opcion == '3':
            print("\nHa elegido la opción 3")
            nombre_pelicula = input("Introduzca el nombre para buscar películas: ")
            if nombre_pelicula in BBDD_peliculas:
                print("Nombre:", nombre_pelicula, "; Año:", datos_pelicula['anio'], "; Género:", datos_pelicula['genero'], "; Duración:", datos_pelicula['duracion'])
            else:
                print("\nNo se encontraron películas.")
    
        elif opcion == '4':
            print("\nHa elegido la opción 4")
            anio_min = int(input("Introduzca la franja de años para buscar películas:\n- Año mínimo: "))
            anio_max = int(input("- Año máximo: "))
    
            peliculas_encontradas = False
            for nombre_pelicula, datos_pelicula in BBDD_peliculas.items():
                if datos_pelicula['anio'] >= anio_min and datos_pelicula['anio'] <= anio_max:
                    peliculas_encontradas = True
                    print("Nombre:", nombre_pelicula, "; Año:", datos_pelicula['anio'], "; Género:", datos_pelicula['genero'], "; Duración:", datos_pelicula['duracion'])
            
            if peliculas_encontradas == False:
                print("\nNo se encontraron películas.")
        
        elif opcion == '5':
            print("\nHa elegido la opción 5")
            genero = input("Introduzca el género para buscar películas: ")
    
            peliculas_encontradas = False
            for nombre_pelicula, datos_pelicula in BBDD_peliculas.items():
                if datos_pelicula['genero'] == genero:
                    peliculas_encontradas = True
                    print("Nombre:", nombre_pelicula, "; Año:", datos_pelicula['anio'], "; Género:", datos_pelicula['genero'], "; Duración:", datos_pelicula['duracion'])
            
            if peliculas_encontradas == False:
                print("\nNo se encontraron películas.")
        
        elif opcion == '6':
            print("\nHa elegido la opción 6")
            duracion_input = int(input("Introduzca la duración para buscar películas: "))
    
            peliculas_encontradas = False
            for nombre_pelicula, datos_pelicula in BBDD_peliculas.items():
                if duracion_input - datos_pelicula['duracion'] <= 60:
                    peliculas_encontradas = True
                    print("Nombre:", nombre_pelicula, "; Año:", datos_pelicula['anio'], "; Género:", datos_pelicula['genero'], "; Duración:", datos_pelicula['duracion'])
        
            if peliculas_encontradas == False:
                print("\nNo se encontraron películas.")
        
        elif opcion == '7':
            print("\n¡Hasta luego!")
            break
    
        else:
            print("\nIntroduce una opción válida.")
```

## <font color='FIREBRICK'>Ejercicio 4 (2 puntos)</font>

<font size='3px'><p style="text-align: justify;line-height: 150%">Para no perder los datos que el usuario va introduciendo vamos a almacenar ambas BD en dos fichero, uno llamado 'actores.txt' y otro 'películas.txt'. De esta forma, cada vez que se arranca el programa deberán recuperarse los datos almacenados previamente y al introducir la opción 1 (introducir nuevos datos) en ambos casos se debe escribir la información en fichero (además de actualizar la varible correspondiente como hasta ahora).</p></font>

```python
# Realice el ejercicio dentro del cuadro a continuación de esta línea 
def guardar_actores():
    fichero = open('actores.txt', 'w')
    
    for nombre_actor, datos_actor in BBDD_actores.items():
        datos_actor_str = nombre_actor + ";" + str(datos_actor['anio']) + ";" + datos_actor['sexo']
        fichero.write(datos_actor_str + '\n')

    fichero.close()

def guardar_peliculas():
    fichero = open('peliculas.txt', 'w')
    
    for nombre_pelicula, datos_pelicula in BBDD_peliculas.items():
        datos_pelicula_str = nombre_pelicula + ";" + str(datos_pelicula['anio']) + ";" + datos_pelicula['genero'] + ";" + str(datos_pelicula['duracion'])
        fichero.write(datos_pelicula_str + '\n')

    fichero.close()

def cargar_actores():
    fichero = open('actores.txt', 'r')
    
    for linea in fichero:
        lista_datos = linea.split(";")
        actor = lista_datos[0]
        anio = int(lista_datos[1])
        sexo = lista_datos[2]
        peliculas = lista_datos[3:]
        BBDD_actores[actor] = {'anio': anio, 'sexo': sexo, 'peliculas': peliculas}
    
    fichero.close()

def cargar_peliculas():
    fichero = open('peliculas.txt', 'r')
    
    for linea in fichero:
        lista_datos = linea.split(";")
        pelicula = lista_datos[0]
        anio = int(lista_datos[1])
        genero = lista_datos[2]
        duracion = int(lista_datos[3])
        BBDD_peliculas[pelicula] = {'anio': anio, 'genero': genero, 'duracion': duracion}

    fichero.close()

def menu_principal(): 
    cargar_actores()
    cargar_peliculas()

    while True:
        print("\n--- Menú principal ---")
        print("1 - BBDD actores")
        print("2 - BBDD películas")
        print("3 - Salir")
        
        opcion = input("Introduzca una opción: ")
        
        if opcion == '1':
            menu_actores()
            guardar_actores()
        
        elif opcion == '2':
            menu_peliculas()
            guardar_peliculas()
            
        elif opcion == '3':
            print("\n¡Hasta luego!")
            break
            
        else:
            print("\nIntroduce una opción válida.")
```

```python
menu_principal()
```
